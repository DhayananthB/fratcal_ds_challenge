```python
import re
def extract_first_emoji(column):

    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  
        u"\U0001F300-\U0001F5FF"  
        u"\U0001F680-\U0001F6FF"  
        u"\U0001F1E0-\U0001F1FF"  
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE)

    def get_first_emoji(text):

        if pd.isna(text):
            return ''
        text = str(text)
        emojis = emoji_pattern.findall(text)
        return emojis[0] if emojis else ''
    
    return column.apply(get_first_emoji)

```

```python
# Extracting emojis 
def extract_emojis(df):
    df['ph'] = extract_first_emoji(df['ph'])
    df['Hardness'] = extract_first_emoji(df['Hardness'])
    df['Solids'] = extract_first_emoji(df['Solids'])
    df['Turbidity'] = extract_first_emoji(df["Turbidity"])
```
